```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import math
import random
```

```python
random.seed(2)
sample_size = 10; sample_min = -2; sample_max = 2;
```

```python
def gaussian(x, mu, sigma):
    return np.exp(np.power((x-mu), 2) / (2 * np.power(sigma, 2))) / np.sqrt(2 * np.pi * np.power(sigma, 2))
```

```python
# Initialise: Randomly initialising the model's parameters.
sampled_points = np.random.uniform(low=sample_min, high=sample_max, size=(sample_size))
mu = np.random.rand(2)
sigma = np.random.rand(2)
prior = 0.5 * np.ones((2))
H = np.zeros((sample_size, 2))
```

```python
# E-step: Using the current model parameters to carry out soft assignments.
def e_step():
    P_Oi_m0 = gaussian(sampled_points, mu[0], sigma[0])
    P_Oi_m1 = gaussian(sampled_points, mu[1], sigma[1])
    H[:, 0] = (P_Oi_m0 * prior[0]) / (P_Oi_m0 * prior[0] + P_Oi_m1 * prior[1])
    H[:, 1] = (P_Oi_m1 * prior[1]) / (P_Oi_m0 * prior[0] + P_Oi_m1 * prior[1])
```

```python
# M-step: Re-estimating the model parameters.
def m_step():
    mu[0] = np.sum(H[:, 0] * sampled_points) / np.sum(H[:, 0])
    mu[1] = np.sum(H[:, 1] * sampled_points) / np.sum(H[:, 1])
    sigma[0] = np.sqrt(np.sum(H[:, 0] * np.power(sampled_points-mu[0],2)) / np.sum(H[:, 0]))
    sigma[1] = np.sqrt(np.sum(H[:, 1] * np.power(sampled_points-mu[1],2)) / np.sum(H[:, 1]))  
    prior[0] = np.mean(H[:, 0])
    prior[1] = np.mean(H[:, 1])
```

```python
# Iterate.
iteration = 1
curr_diff = 1
# while curr_diff > diff_thresold:
while iteration <= 20:
    print "\nIteration no:", iteration
    x_0 = np.linspace(np.min(sampled_points), np.max(sampled_points), 100)
    x_1 = np.linspace(np.min(sampled_points), np.max(sampled_points), 100)
    plt.scatter(sampled_points, np.zeros(sample_size), color='red')
    plt.plot(x_0,mlab.normpdf(x_0, mu[0], sigma[0]), color='green')
    plt.plot(x_1,mlab.normpdf(x_1, mu[1], sigma[1]), color='blue')    
    plt.show()

    e_step()
    m_step()
    
    print "Mean: ", mu[0], ":", mu[1]
    print "Standard Deviation: ", sigma[0], ":", sigma[1]
    print "Prior Probabilities: ", prior[0], ":", prior[1]
    iteration += 1
```
